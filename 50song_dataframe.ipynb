```python
from konlpy.tag import Okt, Komoran, Kkma, Hannanum, Twitter
okt = Okt()

def text_clean(doc):
    text = re.sub('[^가-힣ㄱ-ㅎㅏ-ㅣ ]', '', doc)
    return text
def remove_stwords_oneletter(doc):
  doc = okt.morphs(doc)
  doc = [t for t in doc if len(t) > 1]
  doc = ' '.join(doc)
  return doc
```

```python
tokens = []
for i in song481.가사 :
    i = remove_stwords_oneletter(i)
    i = text_clean(i)
    i = okt.morphs(i)
    tokens.append(i)
```

```python
sent = pd.DataFrame(columns = ['슬픔', '분노', '기쁨', '긴장감', '무서운', '증오', '차분히', '섹시', '브릿락', '설렘', '몽환'])

for i in range(0,481) :
    for tok in tokens[i] : 
        try :
            ser = df.loc[tok]
            sent['기쁨'].append(ser[df.loc[tok] == 1].index[0])
        except:
            pass
```

```python
sent
```

```python
counter={}
for i in range(0,481) :
    for i in tokens[i]:
        for j in df.index:
            if i == j:
                if i in counter:
                    counter[i] += 1
                else:
                    counter[i] = 1
```

```python
counter
```

```python
emotion_score = {}

for key in counter.keys():
    for word in df.index:
        if key == word:
            for label in df.columns[df.loc[key,:] == 1]:
                if label in emotion_score:
                    emotion_score[label] += 1
                else:
                    emotion_score[label] = 1
```

```python
emotion_score
```

```python
for i, word_li in enumerate(tokens):
    for word in word_li:
        if word in df.index:
            emotion = df.columns[df.loc[word] == 1]
            title = song.index[i]
            print(title, emotion)
            song.loc[title, emotion] += 1
```
