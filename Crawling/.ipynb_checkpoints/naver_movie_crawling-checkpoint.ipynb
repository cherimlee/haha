```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://movie.naver.com/movie/point/af/list.naver?&page={}'
res = requests.get(url)
html = res.text
soup = BeautifulSoup(html, 'html.parser')
path = '#old_content'
movie = soup.select(path)
movie
```

```python
for i in movie :
    print("제목:",i.select('a')[0].text)
    print("평점:",i.select('div > em')[0].text)
    print("후기:",i.select('a')[1]['onclick'].split(',')[2])
    print("날짜:",i.select('td.num')[-1].text.split('*')[-1])
    print('-------------------------------------')
```
