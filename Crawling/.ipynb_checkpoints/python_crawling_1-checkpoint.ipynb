```python
url = "https://www.imdb.com/movies-coming-soon/2022-02"
res = requests.get(url)
html = res.text
soup = BeautifulSoup(html, 'html.parser')
path = "#main > div > div.list.detail > div"
movie = soup.select(path)
len(movie)
```

```python
movie[0]
```

```python
for i in movie : 
    print("Title:",i.select('a')[0].text.strip())
    print("Genre:",i.select('span')[0].text, i.select('span')[2].text)
    print("Content:",i.select('div.outline')[0].text.strip())
    print("Director:",i.select('span > a')[0].text)
    print("Stars:",i.select('div>a')[:4])
    print("-"*100)
```
